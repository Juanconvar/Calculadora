```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

G = nx.DiGraph()

edges = [
    ("Casita", "Librería Gauss", 180),
    ("Casita", "Librería Fahrenheit", 314),
    ("Casita", "Librería Fibonacci", 345),
    ("Librería Fibonacci", "Librería Gadner", 310),
    ("Librería Gadner", "Librería Voronoi", 170),
    ("Librería Voronoi", "Librería Gauss", 190),
    ("Librería Gauss", "Librería Richter", 225),
    ("Librería Richter", "Librería Celsius", 280),
    ("Librería Fahrenheit", "Librería Celsius", 255),
    ("Librería Fahrenheit", "Librería Hilbert", 230),
    ("Librería Hilbert", "Librería Celsius", 312),
    ("Librería Voronoi", "Librería Euler", 317),
    ("Librería Euler", "Librería Könisberg", 275),
    ("Librería Gauss", "Librería Könisberg", 198),
    ("Librería Euler", "Librería Gadner", 415),
    ("Librería Euler", "Librería Gauss", 230),
    ("Librería Fibonacci", "Casita", 299),
    ("Librería Gauss", "Casita", 330),
]


G.add_weighted_edges_from(edges)

plt.figure(figsize=(12, 8))
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=3000, font_size=12, font_weight='bold')
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Mapa de Librerías y Caminos")
plt.show()


adj_matrix = nx.adjacency_matrix(G).todense()
adj_df = pd.DataFrame(adj_matrix, index=G.nodes, columns=G.nodes)

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)

print("Matriz de Adyacencia:\n")
print(adj_df)

inc_matrix = nx.incidence_matrix(G, oriented=True).todense()
inc_df = pd.DataFrame(
    inc_matrix, index=G.nodes, columns=[f"Arista {i}" for i in range(1, len(edges) + 1)]
)

print("\nMatriz de Incidencia:\n")
print(inc_df)

origin = "Casita"
destination = "Librería Celsius"
shortest_path = nx.dijkstra_path(G, origin, destination)
shortest_distance = nx.dijkstra_path_length(G, origin, destination)

print(f"\nEl camino más corto de {origin} a {destination} es:")
for i, node in enumerate(shortest_path[:-1]):
    next_node = shortest_path[i + 1]
    weight = G[node][next_node]["weight"]
    print(f"   {node} -> {next_node} (Peso: {weight})")
print(f"Distancia total: {shortest_distance}")

plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=3000, font_size=12, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

shortest_edges = [(shortest_path[i], shortest_path[i + 1]) for i in range(len(shortest_path) - 1)]
nx.draw_networkx_edges(G, pos, edgelist=shortest_edges, edge_color="red", width=3)
plt.title(f"Camino Más Corto: {origin} -> {destination}")
plt.show()
```
